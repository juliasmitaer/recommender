```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
data = pd.read_csv('data.csv')
```

```python
df = data.copy()
pf = pd.DataFrame(df)
question = pf[["question_id","category","user_id"]]
print(question[:5])
```

```python
df2 = data.copy()
pf2 = pd.DataFrame(df2)

pf2['benar'] = np.where(pf2['B_S'] == pf2['answer'], 1, 0)

answers = pf2[["question_id","user_id","benar"]]
answers.head()
```

```python
import tensorflow as tf
from tensorflow import keras
import tensorflow_datasets as tfds
from tensorflow.keras import Model
from tensorflow.keras.layers import Input,Embedding,Dot,Flatten,Dense,LSTM
from tensorflow.keras.optimizers import Adam
```

```python
import numpy.ma as ma
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
import tabulate
```

```python
qs = train_test_split(question,test_size=0.2)
aw = train_test_split(answers,test_size=0.2)

SPLIT_TRAIN = 0.8

train_size = int(SPLIT_TRAIN * len(qs))
train_size_ = int(SPLIT_TRAIN * len(aw))

train_question = qs[:train_size]
test_question = qs[train_size:]

train_answers = aw[:train_size_]
test_answers = aw[train_size_:]
```

```python
EMBEDDING_DIM = 8

# Get the maximum question ID in the entire dataset
max_question_id = max(max(train_answers[0]['question_id'].values), max(test_answers[0]['question_id'].values) if test_answers[0].size > 0 else 0)
max_user_id = max(max(train_answers[0]['user_id'].values), max(test_answers[0]['user_id'].values) if test_answers[0].size > 0 else 0)

question_input = Input(shape=(1,))
user_input = Input(shape=(1,))

question_Embedding = Embedding(max_question_id+1,output_dim=EMBEDDING_DIM)(question_input)
user_Embedding = Embedding(max_user_id+1,output_dim=EMBEDDING_DIM)(user_input)

question_LSTM = LSTM(12)(question_Embedding)
user_LSTM = LSTM(12)(user_Embedding)

question_flatten = Flatten()(question_LSTM)
user_flatten = Flatten()(user_LSTM)

question_dense = Dense(10)(question_flatten)
user_dense = Dense(10)(user_flatten)

result = Dot(axes=1)([question_dense,user_dense])

model = Model([question_input,user_input],result)

model.summary()

model.compile(optimizer='adam',loss='mse')
```

```python
question_ids = train_answers[0]['question_id'].values
user_ids = train_answers[0]['user_id'].values
benar_values = train_answers[0]['benar'].values

model.fit(x=[question_ids,user_ids], y=benar_values, epochs=30)
```

```python
model.evaluate(x=[test_answers[0]['question_id'].values,test_answers[0]['user_id'].values],y=test_answers[0]['benar'].values)
```

```python
prediction = model.predict(x=[test_answers[0]['question_id'].values,test_answers[0]['user_id'].values])
```
